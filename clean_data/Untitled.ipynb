```python
import pandas as pd
import nltk
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from tqdm import tqdm
import numpy as np
from hazm import *
from __future__ import unicode_literals
```

```python
df = pd.read_excel("../data/eshia_detect_lang.xlsx")
```

```python
data = df[df["Language"] == "fa"]
```

```python
normalizer = Normalizer()
```

```python
data["Tokenize data"] = np.empty((len(data), 0)).tolist()
```

```python
data.reset_index(inplace=True, drop=True)
```

```python
data = df
```

```python
word_tokenize(data["Book name"][0] + " " + data["Book content"][0])
```

```python
for i in range(len(data)):
    tokens = word_tokenize(data["Book name"][i] + " " + data["Book content"][i])
    data["Tokenize data"][i] = tokens
```

```python
lr = LogisticRegression(max_iter=1000, multi_class='multinomial', verbose=True,)
```

```python
lr = LogisticRegression(penalty='l2', max_iter=1000,multi_class='multinomial', solver='sag', verbose=True,l1_ratio=0.03)
```

```python
print('Precision on Test Set: ', lr.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix,\
    f1_score
def eval_results(y_true, y_pred):
    print(classification_report(y_true, y_pred))
    print(confusion_matrix(y_true, y_pred))
    print(f1_score(y_true, y_pred, average='weighted'))
```

```python
eval_results(y_test, lr.predict(X_test))
```
